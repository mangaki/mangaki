```python
import os
os.environ.setdefault("DJANGO_SETTINGS_MODULE", "mangaki.settings")

import django
django.setup()

from django.conf import settings
```

```python
from mangaki.utils.data import Dataset

data = Dataset()
data.load_csv('ratings.csv', convert=str)
```

```python
from mangaki.models import Work
import numpy as np
import csv

"""dataset.titles = {}
for work_id, title in Work.objects.values_list('id', 'title'):
    dataset.titles[work_id] = title"""

with open(os.path.join(settings.DATA_DIR, 'works.csv')) as f:
    titles = [title for _, title in csv.reader(f)]
    titles = np.array(titles, dtype=np.object)
```

```python
from collections import Counter
import pandas as pd

nb_ratings = Counter()
nb_liked = Counter()
nb_disliked = Counter()
for (user_id, work_id), choice in zip(data.anonymized.X, data.anonymized.y):
    nb_ratings[work_id] += 1
    if choice in ['like', 'favorite']:
        nb_liked[work_id] += 1
    elif choice in ['neutral', 'dislike']:
        nb_disliked[work_id] += 1
total_ratings = sum(nb_ratings.values())
total_liked = sum(nb_liked.values())
total_disliked = sum(nb_disliked.values())

ratings = 0
liked = 0
disliked = 0
lines = []
for rank, (work_id, nb) in enumerate(nb_ratings.most_common(100), start=1):
    ratings += nb_ratings[work_id]
    liked += nb_liked[work_id]
    disliked += nb_disliked[work_id]
    lines.append({
        'title': titles[work_id],
        'nb_ratings': nb,
        '% of ratings': ratings * 100 / total_ratings,
        '% of likes': liked * 100 / total_liked,
        '% of dislikes': disliked * 100 / total_disliked
    })
df = pd.DataFrame(lines)
df.index += 1
df
```
